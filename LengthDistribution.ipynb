```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (11, 8)
length_df = pd.read_csv("/Users/mjohnson/Desktop/Projects/AngiospermHybSeq/length_distribution.txt",sep="\t",header=None)
length_df.columns = ["Gene","Clade","PctLength"]
clade_order = ["ANA_grade","calamus","alismatids","pandan_yams","liliids","commelinids","coontails","sarcandrae","magnoliids","ranunculids","proteas_sl","wheel_tree","boxwood","giant_rhubarb","coral_plant","grapevines","mistletoes_sl","saxifrages_sl","guinea_flowers","early_caryophyllids","amaranths_sl","cacti_sl","four_o_clocks_sl","bladdernuts","cranesbills_sl","gourds_sl","roses_sl","oaks_sl","legumes_sl","caltrops","myrtils_sl","bittersweets","olomea","spurges_sl","sourgrasses_sl","soapberries_sl","mallows_sl","crucifers_sl","dogwoods_sl","heaths_sl","umbelliferae_sl","daisies_sl","tassel_bushes","hollies","monkey_chair","nightshades_sl","forget_me_not","gentians_sl","sages_sl"]
```

```python
ax = length_df.boxplot("PctLength",by="Clade")
rotations  = ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=90)
```

```python
ax = sns.stripplot(x="Clade", y="PctLength", data=length_df, jitter=True,order=clade_order,size=8)
rotations  = ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=90)
```
